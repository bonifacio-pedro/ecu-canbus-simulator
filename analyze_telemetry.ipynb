```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('telemetry_log.csv')
```

```python
# Plot
fig, axs = plt.subplots(4, 1, figsize=(14, 12))

# RPM
axs[0].plot(df['Timestamp']/1000, df['RPM'], color='red', linewidth=2)
axs[0].set_ylabel('RPM')
axs[0].grid(True, alpha=0.3)

# Speed
axs[1].plot(df['Timestamp']/1000, df['Speed'], color='blue', linewidth=2)
axs[1].set_ylabel('Speed (km/h)')
axs[1].grid(True, alpha=0.3)

# Throttle vs Brake
axs[2].plot(df['Timestamp']/1000, df['Throttle'], label='Throttle', color='green')
axs[2].plot(df['Timestamp']/1000, df['BrakePressure'], label='Brake', color='red')
axs[2].set_ylabel('%')
axs[2].legend()
axs[2].grid(True, alpha=0.3)

# Engine Temp
axs[3].plot(df['Timestamp']/1000, df['EngineTemp'], color='orange', linewidth=2)
axs[3].axhline(y=110, color='r', linestyle='--', label='Overheating')
axs[3].set_ylabel('Temp (C)')
axs[3].legend()
axs[3].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('ecu_telemetry_analysis.png', dpi=300)
plt.show()

print(f" Max RPM: {df['RPM'].max()}")
print(f" Max Speed: {df['Speed'].max()} km/h")
print(f" Max Temp: {df['EngineTemp'].max()}C")
```
